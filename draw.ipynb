```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# learning rate R8
x = [1, 2, 3, 4, 5]
lr = [0.05, 0.02, 0.01, 0.005, 0.002]
acc = [0.9689, 0.9703, 0.9712, 0.9721, 0.9708]
epoch = [61, 98, 161, 200, 200]
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111)
lns1 = ax1.plot(x, acc, 'm.-.', label='acc')



ax2 = ax1.twinx()
lns2 = ax2.plot(x, epoch, 'c*-', label='epoch')

lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax2.legend(lns, labs)


ax1.grid()
plt.xticks(x, lr)
ax1.set_ylabel("Accuracy")
ax2.set_ylabel("Number of epoch")
ax1.set_xlabel("Learning rate")
ax1.set_ylim([0.93, 1.0])
# plt.show()
plt.savefig('lr.png')
```

```python
# embedding size R8 and R52

x = [0, 1, 2, 3, 4, 5, 6]
size = [10, 50, 100, 150, 200, 250, 300]
acc_r8 = [0.9497, 0.9612, 0.9607, 0.9630, 0.9639, 0.9662, 0.9653]
acc_r52 = [0.8750, 0.9381, 0.9354, 0.9373, 0.9373, 0.9404, 0.9385]
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111)

ax1.plot(x, acc_r8, 'co-.', label='R8')
ax1.plot(x, acc_r52, 'b*-.', label='R52')
ax1.legend()


ax1.grid()
plt.xticks(x, size)
ax1.set_ylabel("Accuracy")
ax1.set_xlabel("Embedding size")

plt.savefig('embedding.png')
```
